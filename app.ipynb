```python
track_count_by_genre = chinook_df.copy()
track_count_by_genre = (
        track_count_by_genre
        .groupby('genre_name')
        .agg(num_tracks=('track_id' , 'count'))
        .reset_index()
    )
sns.barplot(
    data=track_count_by_genre.sort_values(by='num_tracks',ascending=False),
    x='num_tracks',
    y='genre_name',
    palette='Set2'
)
plt.show()
```
